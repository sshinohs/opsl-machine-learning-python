```python
import os
import numpy as np

import import_ipynb
import khoa

import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

code

```python
dir_path = '../data/buoy'
```

```python
file_path_list = khoa.path_khoa(dir_path)
```

```python
data_khoa = np.empty(shape=(0,8))
```

```python
date_time = np.empty(shape=(0), dtype='datetime64')
```

```python
for file_path in file_path_list:
    header, date_time_single, data = khoa.load_khoa(file_path)
    data_khoa = np.append(data_khoa, data, axis=0)
    np.array(date_time_single)
    date_time = np.concatenate((date_time, date_time_single), axis=0)
```

```python
plt.figure(figsize=(10,5))
plt.plot(date_time, data_khoa[:,6], '.-')
plt.xlabel('series')
plt.ylabel('current velocity(cm/s)')
plt.grid(True)
plt.show()
```
